#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python

# --------------------------------
# 1. Clean column names
# --------------------------------
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
)

# --------------------------------
# 2. Group by gender (numeric only)
# --------------------------------
gender_group = df.groupby('gender').mean(numeric_only=True)

# --------------------------------
# 3. Automatically find math column
# --------------------------------
math_col = [col for col in gender_group.columns if 'math' in col][0]

# --------------------------------
# 4. Compute total average (mean of all numeric scores)
# --------------------------------
gender_group['total_average'] = gender_group.mean(axis=1)

# --------------------------------
# 5. Prepare data for plotting
# --------------------------------
labels = ['Total Average', 'Math Average']
x = np.arange(len(labels))

female_scores = [
    gender_group.loc['female', 'total_average'],
    gender_group.loc['female', math_col]
]

male_scores = [
    gender_group.loc['male', 'total_average'],
    gender_group.loc['male', math_col]
]

# --------------------------------
# 6. Plot
# --------------------------------
plt.figure(figsize=(10, 8))

plt.bar(x - 0.2, male_scores, 0.4, label='Male')
plt.bar(x + 0.2, female_scores, 0.4, label='Female')

plt.xticks(x, labels)
plt.ylabel("Marks")
plt.title(
    "Total Average vs Math Average Marks by Gender",
    fontweight='bold'
)
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
